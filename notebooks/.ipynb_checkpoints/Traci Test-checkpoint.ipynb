```python
def run_day_sim(n_steps):
    sumoCmd = [sumoBinary, "-c", "../sumo_simulation/sim_config/km2_jesusmaria/scenario/osm.sumocfg"]
    subs = []
    traci.start(sumoCmd)
    
    id_list = traci.edge.getIDList()
#     for edge_id in traci.edge.getIDList():
#         traci.edge.subscribe(edge_id,[traci.constants.VAR_SPEED])
    
#     for edge_id in traci.lane.getIDList():
#         traci.lane.subscribe(edge_id, [traci.constants.VAR_CO2EMISSION,
#                                        traci.constants.VAR_COEMISSION,
#                                        traci.constants.VAR_PMXEMISSION,
#                                        traci.constants.VAR_NOXEMISSION,
#                                        traci.constants.VAR_NOISEEMISSION,
#                                        traci.constants.VAR_FUELCONSUMPTION])
    
    traci.simulationStep(500)
    subs = [traci.edge.getLastStepMeanSpeed(edge_id) for edge_id in id_list]
    subs_co2 = [traci.edge.getCO2Emission(edge_id) for edge_id in id_list]
    #subs.append(traci.lane.getAllSubscriptionResults())
    
    #subscriptions = traci.lane.getAllSubscriptionResults()
    traci.close(False)
    return subs, subs_co2
```
