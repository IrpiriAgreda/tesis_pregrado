```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
with open('output/rewards_experiment_2019-06-10 10:03:46611928.csv') as f:
    text = f.read()
```

```python
df = pd.DataFrame(text.split(r'\n'))
```

```python
for index, row in df.iterrows():
    line = row[0].split(',')
    try:
        df.loc[index, 'episode'] = line[0]
        df.loc[index, 'timestep'] = line[1]
        df.loc[index, 'reward'] = line[2]
    except:
        pass
```

```python
df = df.dropna()
```

```python
del df[0]
```

```python
for column in df.columns:
    df[column] = df[column].apply(float)
```

```python
avg_per_episode = df[['episode', 'reward']].groupby('episode').mean()
```

```python
avg_per_episode.plot()
plt.title('Average Training Reward per Episode')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.savefig('avg_per_episode_v1.png', dpi=300)
plt.show()
```

```python
df['reward'].plot()
plt.title('Training Reward per Timestep')
plt.xlabel('Timestep')
plt.ylabel('Reward')
plt.savefig('avg_per_timestep_v1.png', dpi=300)
plt.show()
```
